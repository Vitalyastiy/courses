```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Создаем тестовый набор данных
data = {'Apples': [30, 20, 25, 15],
        'Oranges': [20, 25, 30, 10],
        'Pears': [10, 15, 10, 30],
        'Bananas': [25, 30, 15, 20]}
df = pd.DataFrame(data, index=['Spring', 'Summer', 'Fall', 'Winter'])

# Линейный график
df.plot(kind='line')
plt.title('Line Plot')
plt.show()

# Столбчатая диаграмма
df.plot(kind='bar')
plt.title('Bar Plot')
plt.show()

# Гистограмма
df.plot(kind='hist', alpha=0.5)
plt.title('Histogram Plot')
plt.show()

# Ящик с усами (Box Plot)
df.plot(kind='box')
plt.title('Box Plot')
plt.show()

# Круговая диограмма
df.plot(kind='pie', subplots=True, legend=False)
plt.title('Pie Plot')
plt.show()

# Scatter plot
df.plot(kind='scatter', x='Apples', y='Oranges')
plt.title('Scatter Plot')
plt.show()

# Area plot
df.plot(kind='area', stacked=False)
plt.title('Area Plot')
plt.show()

# Heatmap
plt.imshow(df, cmap='hot', interpolation='nearest')
plt.title('Heatmap')
plt.show()
```

Примеры простых визуализаций в пандас

```python
#Лайнплот
data.groupby(['class'], as_index = False).agg({'age':['count','nunique']}).plot(kind = 'line')
```

```python
#Гистограмма
data.groupby(['class'], as_index = False).agg({'age':['count','nunique']}).plot(kind = 'hist')
```

```python
#Барплот + траспонирование
t = data.groupby(['sex', 'embark_town'], as_index = False)['age'].median()
t_p = t.pivot(index='embark_town', columns='sex', values='age')
display(t_p)
t_p.plot(kind = 'bar')
```
